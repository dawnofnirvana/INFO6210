4.What posts are like mine?
We group all the tweets that have same tags. ALl tags in one group could be considered as the similar posts with each other.

```python
from bson.son import SON
pipeline = ([{"$unwind":"$screen_name"},
	{"$group":{"_id":{"text":"$text","tag":"$tag"}}},
	{"$project":{"tagname":"$_id.tag","subinfo":{"text":"$_id.text"}}},
	{"$group":{"_id":"$tagname","info":{"$push":"$subinfo"},
	"infoNo":{"$sum":1}}},{"$sort":{"infoNo":-1}}])
import pprint
pprint.pprint(list(db.posts4.aggregate(pipeline)))
```

5.What users post like me?
Basically, we count how many times each tag have been used by different users. Then we sort all of the users based on the count number. Therefor, the top five users are very similar with each other because they have used the same tag for most of times in their posts.

```python
from bson.son import SON
pipeline = ([{"$unwind":"$screen_name"},
	{"$group":{"_id":{"username":"$screen_name","tag":"$tag"},"count":{"$sum":1}}},
	{"$match":{"count":{"$gt":5}}},{"$sort":{"count":-1}},
	{"$project":{"tagname":"$_id.tag","subinfo":{"User":"$_id.username","sumcount":"$count"}}},
	{"$group":{"_id":"$tagname","info":{"$push":"$subinfo"},
	"infoNo":{"$sum":1}}},{"$sort":{"infoNo":-1}}])
import pprint
pprint.pprint(list(db.posts.aggregate(pipeline)))
```
